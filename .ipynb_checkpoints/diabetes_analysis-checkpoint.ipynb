# تحليل بيانات مرضى السكري
## Diabetes Patients Data Analysis

### 1. استيراد المكتبات Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager
import warnings
warnings.filterwarnings('ignore')

# إعدادات الرسوم البيانية
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

%matplotlib inline
```

### 2. قراءة البيانات Load Data

```python
# قراءة ملف Excel
df = pd.read_excel('diabetes_patients_data.xlsx')

# عرض أول 10 صفوف
print('أول 10 صفوف من البيانات:')
df.head(10)
```

### 3. نظرة عامة على البيانات Data Overview

```python
print('='*60)
print('نظرة عامة على البيانات')
print('='*60)
print(f'عدد السجلات: {len(df)}')
print(f'عدد الأعمدة: {len(df.columns)}')
print(f'
الأعمدة:')
for i, col in enumerate(df.columns, 1):
    print(f'  {i}. {col}')

print('
' + '='*60)
print('معلومات البيانات:')
print('='*60)
df.info()
```

```python
# إحصائيات وصفية
print('إحصائيات وصفية للأعمدة الرقمية:')
df.describe()
```

### 4. تحليل توزيع مرضى السكري Diabetes Distribution

```python
# عدد ونسبة المصابين وغير المصابين
diabetes_counts = df['مصاب بالسكري'].value_counts()
diabetes_percent = df['مصاب بالسكري'].value_counts(normalize=True) * 100

print('='*60)
print('توزيع مرضى السكري')
print('='*60)

for status in diabetes_counts.index:
    count = diabetes_counts[status]
    percent = diabetes_percent[status]
    print(f'{status}: {count} مريض ({percent:.1f}%)')

print(f'
إجمالي المرضى: {len(df)}')
```

```python
# رسم بياني دائري
plt.figure(figsize=(8, 8))
colors = ['#ff9999', '#66b3ff']
diabetes_counts.plot(kind='pie', autopct='%1.1f%%', colors=colors, 
                    startangle=90, textprops={'fontsize': 14})
plt.title('توزيع مرضى السكري', fontsize=18, fontweight='bold', pad=20)
plt.ylabel('')
plt.legend(['غير مصاب', 'مصاب'], loc='best', fontsize=12)
plt.tight_layout()
plt.show()
```

### 5. تحليل العمر Age Analysis

```python
plt.figure(figsize=(12, 6))

# توزيع العمر
plt.subplot(1, 2, 1)
plt.hist(df['العمر'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('توزيع العمر', fontsize=14, fontweight='bold')
plt.xlabel('العمر')
plt.ylabel('التكرار')
plt.grid(True, alpha=0.3)

# صندوق العمر حسب الإصابة بالسكري
plt.subplot(1, 2, 2)
df.boxplot(column='العمر', by='مصاب بالسكري', ax=plt.gca())
plt.title('توزيع العمر حسب الإصابة بالسكري')
plt.suptitle('')  # إزالة العنوان الافتراضي
plt.xlabel('مصاب بالسكري')
plt.ylabel('العمر')

plt.tight_layout()
plt.show()
```

```python
# إحصائيات العمر حسب الإصابة
print('='*60)
print('إحصائيات العمر حسب الإصابة بالسكري')
print('='*60)
age_by_diabetes = df.groupby('مصاب بالسكري')['العمر'].describe()
print(age_by_diabetes)
```

### 6. تحليل مستوى الجلوكوز Glucose Analysis

```python
plt.figure(figsize=(12, 6))

# توزيع الجلوكوز
plt.subplot(1, 2, 1)
plt.hist(df['مستوى الجلوكوز'], bins=20, color='lightcoral', edgecolor='black', alpha=0.7)
plt.title('توزيع مستوى الجلوكوز', fontsize=14, fontweight='bold')
plt.xlabel('مستوى الجلوكوز')
plt.ylabel('التكرار')
plt.axvline(x=100, color='red', linestyle='--', label='الحد الطبيعي (100)')
plt.axvline(x=126, color='darkred', linestyle='--', label='حد السكري (126)')
plt.legend()
plt.grid(True, alpha=0.3)

# مقارنة الجلوكوز حسب الإصابة
plt.subplot(1, 2, 2)
df.boxplot(column='مستوى الجلوكوز', by='مصاب بالسكري', ax=plt.gca())
plt.title('مستوى الجلوكوز حسب الإصابة')
plt.suptitle('')
plt.xlabel('مصاب بالسكري')
plt.ylabel('مستوى الجلوكوز')

plt.tight_layout()
plt.show()
```

```python
# متوسط الجلوكوز حسب الإصابة
print('='*60)
print('متوسط مستوى الجلوكوز حسب الإصابة بالسكري')
print('='*60)
glucose_stats = df.groupby('مصاب بالسكري')['مستوى الجلوكوز'].agg(['mean', 'std', 'min', 'max'])
print(glucose_stats.round(2))
```

### 7. تحليل مؤشر كتلة الجسم (BMI) BMI Analysis

```python
plt.figure(figsize=(12, 6))

# توزيع BMI
plt.subplot(1, 2, 1)
plt.hist(df['مؤشر كتلة الجسم (BMI)'], bins=20, color='lightgreen', edgecolor='black', alpha=0.7)
plt.title('توزيع مؤشر كتلة الجسم (BMI)', fontsize=14, fontweight='bold')
plt.xlabel('BMI')
plt.ylabel('التكرار')
plt.axvline(x=25, color='orange', linestyle='--', label='زيادة الوزن (25)')
plt.axvline(x=30, color='red', linestyle='--', label='سمنة (30)')
plt.legend()
plt.grid(True, alpha=0.3)

# مقارنة BMI حسب الإصابة
plt.subplot(1, 2, 2)
df.boxplot(column='مؤشر كتلة الجسم (BMI)', by='مصاب بالسكري', ax=plt.gca())
plt.title('BMI حسب الإصابة بالسكري')
plt.suptitle('')
plt.xlabel('مصاب بالسكري')
plt.ylabel('BMI')

plt.tight_layout()
plt.show()
```

```python
# تصنيف BMI
def classify_bmi(bmi):
    if bmi < 18.5:
        return 'نحيف'
    elif bmi < 25:
        return 'طبيعي'
    elif bmi < 30:
        return 'زيادة وزن'
    else:
        return 'سمنة'

df['تصنيف BMI'] = df['مؤشر كتلة الجسم (BMI)'].apply(classify_bmi)

print('='*60)
print('توزيع فئات BMI حسب الإصابة بالسكري')
print('='*60)
bmi_diabetes = pd.crosstab(df['تصنيف BMI'], df['مصاب بالسكري'])
print(bmi_diabetes)
print('
النسب المئوية:')
print(pd.crosstab(df['تصنيف BMI'], df['مصاب بالسكري'], normalize='index') * 100)
```

### 8. تحليل ضغط الدم Blood Pressure Analysis

```python
plt.figure(figsize=(12, 6))

# توزيع ضغط الدم
plt.subplot(1, 2, 1)
plt.hist(df['ضغط الدم'], bins=20, color='plum', edgecolor='black', alpha=0.7)
plt.title('توزيع ضغط الدم', fontsize=14, fontweight='bold')
plt.xlabel('ضغط الدم')
plt.ylabel('التكرار')
plt.axvline(x=120, color='orange', linestyle='--', label='طبيعي مرتفع (120)')
plt.axvline(x=140, color='red', linestyle='--', label='ضغط مرتفع (140)')
plt.legend()
plt.grid(True, alpha=0.3)

# مقارنة ضغط الدم حسب الإصابة
plt.subplot(1, 2, 2)
df.boxplot(column='ضغط الدم', by='مصاب بالسكري', ax=plt.gca())
plt.title('ضغط الدم حسب الإصابة')
plt.suptitle('')
plt.xlabel('مصاب بالسكري')
plt.ylabel('ضغط الدم')

plt.tight_layout()
plt.show()
```

### 9. مصفوفة الارتباط Correlation Matrix

```python
# تحويل العمود النصي إلى رقمي للتحليل
df['مصاب_رقمي'] = df['مصاب بالسكري'].map({'لا': 0, 'نعم': 1})

# حساب مصفوفة الارتباط
numeric_cols = ['العمر', 'مستوى الجلوكوز', 'ضغط الدم', 'مؤشر كتلة الجسم (BMI)', 'مصاب_رقمي']
corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, fmt='.2f', cbar_kws={"shrink": .8})
plt.title('مصفوفة الارتباط بين المتغيرات', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()
```

```python
print('='*60)
print('معاملات الارتباط مع الإصابة بالسكري')
print('='*60)
correlations = corr_matrix['مصاب_رقمي'].sort_values(ascending=False)[:-1]
for var, corr in correlations.items():
    print(f'{var}: {corr:.3f}')
```

### 10. تحليل متعدد المتغيرات Multi-variable Analysis

```python
plt.figure(figsize=(14, 10))

# مخطط مبعثر: الجلوكوز vs BMI
plt.subplot(2, 2, 1)
colors = df['مصاب بالسكري'].map({'لا': 'blue', 'نعم': 'red'})
plt.scatter(df['مؤشر كتلة الجسم (BMI)'], df['مستوى الجلوكوز'], c=colors, alpha=0.6, s=50)
plt.xlabel('BMI')
plt.ylabel('مستوى الجلوكوز')
plt.title('الجلوكوز vs BMI')
plt.legend(['غير مصاب', 'مصاب'])
plt.grid(True, alpha=0.3)

# مخطط مبعثر: العمر vs الجلوكوز
plt.subplot(2, 2, 2)
plt.scatter(df['العمر'], df['مستوى الجلوكوز'], c=colors, alpha=0.6, s=50)
plt.xlabel('العمر')
plt.ylabel('مستوى الجلوكوز')
plt.title('الجلوكوز vs العمر')
plt.grid(True, alpha=0.3)

# مخطط مبعثر: العمر vs BMI
plt.subplot(2, 2, 3)
plt.scatter(df['العمر'], df['مؤشر كتلة الجسم (BMI)'], c=colors, alpha=0.6, s=50)
plt.xlabel('العمر')
plt.ylabel('BMI')
plt.title('BMI vs العمر')
plt.grid(True, alpha=0.3)

# مخطط مبعثر: ضغط الدم vs الجلوكوز
plt.subplot(2, 2, 4)
plt.scatter(df['ضغط الدم'], df['مستوى الجلوكوز'], c=colors, alpha=0.6, s=50)
plt.xlabel('ضغط الدم')
plt.ylabel('مستوى الجلوكوز')
plt.title('الجلوكوز vs ضغط الدم')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

### 11. ملخص النتائج Summary of Findings

```python
print('='*60)
print('ملخص تحليل بيانات مرضى السكري')
print('='*60)

# إحصائيات عامة
total = len(df)
diabetic_count = (df['مصاب بالسكري'] == 'نعم').sum()
non_diabetic_count = (df['مصاب بالسكري'] == 'لا').sum()

print(f'1. إجمالي عدد المرضى: {total}')
print(f'2. عدد المصابين بالسكري: {diabetic_count} ({diabetic_count/total*100:.1f}%)')
print(f'3. عدد غير المصابين: {non_diabetic_count} ({non_diabetic_count/total*100:.1f}%)')

# متوسطات المجموعات
print('
4. متوسطات المؤشرات الحيوية:')
print('-'*40)

avg_by_status = df.groupby('مصاب بالسكري')[['العمر', 'مستوى الجلوكوز', 
                                               'ضغط الدم', 'مؤشر كتلة الجسم (BMI)']].mean()
print(avg_by_status.round(2))

# أقوى عوامل الارتباط
print('
5. أقوى عوامل الارتباط بالإصابة بالسكري:')
print('-'*40)
top_correlations = corr_matrix['مصاب_رقمي'].abs().sort_values(ascending=False)[1:4]
for var, corr in top_correlations.items():
    print(f'   - {var}: {corr:.3f}')

print('
' + '='*60)
print('نهاية التحليل')
print('='*60)
```
